```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random
np.random.seed(3)
mean = 35
sigma=15
number_of_entries = 10000
data = np.random.normal(mean, sigma, number_of_entries)
plt.hist(data, bins=100, range=(0,50), color="red")
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
n = 100000

x = np.random.standard_normal(n)
y = 2.0 * x + 4.0 * np.random.standard_normal(n)
plt.hist2d(x,y,bins=75, cmap=plt.cm.plasma)
plt.colorbar()
plt.xlabel('Channel 1')
plt.ylabel('Channel 2')

plt.show()
```
